# House Sales in King County, USA


This dataset contains house sale prices for King County, which includes Seattle. It includes homes sold between May 2014 and May 2015.


| Variable      | Description                                                                                                 |
| ------------- | ----------------------------------------------------------------------------------------------------------- |
| id            | A notation for a house                                                                                      |
| date          | Date house was sold                                                                                         |
| price         | Price is prediction target                                                                                  |
| bedrooms      | Number of bedrooms                                                                                          |
| bathrooms     | Number of bathrooms                                                                                         |
| sqft_living   | Square footage of the home                                                                                  |
| sqft_lot      | Square footage of the lot                                                                                   |
| floors        | Total floors (levels) in house                                                                              |
| waterfront    | House which has a view to a waterfront                                                                      |
| view          | Has been viewed                                                                                             |
| condition     | How good the condition is overall                                                                           |
| grade         | overall grade given to the housing unit, based on King County grading system                                |
| sqft_above    | Square footage of house apart from basement                                                                 |
| sqft_basement | Square footage of the basement                                                                              |
| yr_built      | Built Year                                                                                                  |
| yr_renovated  | Year when house was renovated                                                                               |
| zipcode       | Zip code                                                                                                    |
| lat           | Latitude coordinate                                                                                         |
| long          | Longitude coordinate                                                                                        |
| sqft_living15 | Living room area in 2015(implies-- some renovations) This might or might not have affected the lotsize area |
| sqft_lot15    | LotSize area in 2015(implies-- some renovations)                                                            |

# 2: Data Wrangling


Drop the columns <code>"id"</code>  and <code>"Unnamed: 0"</code> from axis 1 then  obtain a statistical summary of the data

```python
df.drop(['id','Unnamed: 0'],axis=1,inplace=True)
df.describe()
```

We can see we have missing values for the columns 


```python
df.isnull().sum()
```

```python
print("number of NaN values for the column bedrooms :", df['bedrooms'].isnull().sum())
print("number of NaN values for the column bathrooms :", df['bathrooms'].isnull().sum())
```

We can replace the missing values of the column <code>'bedrooms'</code> with the mean of the column  <code>'bedrooms' 

```python
mean=df['bedrooms'].mean()
df['bedrooms'].replace(np.nan,mean, inplace=True)
```

We also replace the missing values of the column <code>'bathrooms'</code> with the mean of the column  <code>'bathrooms' 

```python
mean1=df['bathrooms'].mean()
df['bathrooms'].replace(np.nan,mean1, inplace=True)
```

```python
print("number of NaN values for the column bedrooms :", df['bedrooms'].isnull().sum())
print("number of NaN values for the column bathrooms :", df['bathrooms'].isnull().sum())
```

# 4: Model Development

```python
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
```

```python
X = df[["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]  ]
Y = df['price']
```

```python
lm1 = LinearRegression()
lm1.fit(X,Y)
```

```python
lm1.intercept_
```

```python
lm1.coef_
```

```python
yhat1 = lm1.predict(X)
yhat1[0:5]
```

# 5: Model Evaluation


<code>train_test_split</code> evaluation approch has been used for model evaluation

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 4)
print('train set:',X_train.shape,Y_train.shape)
print('Test set:', X_test.shape,Y_test.shape)
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,Y_train)
```

<code>r^2</code> value for test data 

```python
#r^2 value for test data
lm.score(X_test,Y_test)
```

<code>r^2</code> value for training data 

```python
#r^2 value for train data 
lm.score(X_train,Y_train)
```

```python
#prediction using train data 
yhat_train = lm.predict(X_train)
yhat_train[0:5]
```

```python
#prediction using test data 
yhat_test = lm.predict(X_test)
yhat_test[0:5]
```

```python
def DistributionPlot (redfunction,bluefunction,redname,bluename,Title):
    plt.figure(figsize=(12,10))
    ax1 = sns.distplot(redfunction,hist=False,color='r', label = redname)
    ax2 = sns.distplot(bluefunction,hist=False,color='b', label = bluename,ax=ax1)
    plt.title(Title)
    plt.xlabel('Price')
    plt.ylabel('features of house')
    plt.show()
    plt.close()
```

```python
Title = 'distribution plot of predicted value using test data vs data distribution of test data'
DistributionPlot(Y_test,yhat_test,'Atcual values(test)','Predicted values(test)',Title)
```

From above : <code>r^2</code> value for test data and training data are equal therefore LinerRegression model for prediction of <code>price</code> for house is acceptable
